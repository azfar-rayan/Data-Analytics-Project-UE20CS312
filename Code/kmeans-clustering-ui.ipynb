```python
df = pd.read_csv('D:\\Divya\\Sem 5\\DA\\Project\\MCI_2021.csv')
```

```python
df["Location"] = df['Latitude'].astype(str) +","+ df["Longitude"].astype(str)
df['Location'].head()
```

```python
from sklearn.cluster import KMeans
x = df[['Longitude', 'Latitude']]
x
```

```python
import gradio as gr
```

```python
x = df[['Longitude', 'Latitude']]

def get_kmeans(y):
    kmeans = KMeans(y+1)        #y clusters will be formed
    kmeans.fit(x)
    #predict cluster
    identified_clusters = kmeans.fit_predict(x)
    data_with_clusters = df.copy()
    data_with_clusters['Cluster'] = identified_clusters
    fig=plt.figure(figsize=(18,18))
    plt.scatter(df['Longitude'], df['Latitude'], c=data_with_clusters['Cluster'], cmap = 'rainbow', alpha=0.5)
    plt.xlim(-79,-79.9)
    plt.ylim(43.3, 44)
    return fig

#input_model = plt.scatter(df['Longitude'], df['Latitude'], c=data_with_clusters['Cluster'], cmap = 'rainbow', alpha=0.5)
#plot=gr.Plot(input_model)
iface = gr.Interface(fn=get_kmeans, inputs=gr.inputs.Slider(1,20, label="No. of Clusters"), outputs=['plot'],title="KMeans clusters of the city")
iface.launch(share=True)
```
